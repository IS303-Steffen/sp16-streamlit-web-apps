

### Chapter Summary

Streamlit is a Python library that turns regular Python scripts into shareable web apps—perfect for business dashboards and interactive tools.

Key concepts include:
- Installing and launching Streamlit
- Using `st.write` and other layout/text functions
- Adding widgets like buttons, inputs, and selectboxes
- Understanding how Streamlit re-runs your script top to bottom every time a user interacts with a widget
- Using `st.session_state` to persist information across reruns
